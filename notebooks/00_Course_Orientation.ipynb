## 6. Quick Python/NumPy Refresher

A minimal sanity check. If any of this feels unfamiliar, work through `DL050` before proceeding.

````python
# --- Python basics ---
# List comprehension
squares = [x**2 for x in range(10)]
print("Squares:", squares)

# Dictionary comprehension
sq_dict = {x: x**2 for x in range(5)}
print("Dict:   ", sq_dict)

# Lambda + map
doubled = list(map(lambda x: x * 2, range(5)))
print("Doubled:", doubled)
````

````python
import numpy as np

np.random.seed(42)  # reproducibility

# --- NumPy basics ---
a = np.array([1, 2, 3, 4, 5])
b = np.random.randn(5)  # 5 samples from standard normal

print("a      :", a)
print("b      :", b.round(3))
print("a + b  :", (a + b).round(3))
print("dot(a,b):", np.dot(a, b).round(3))
````

````python
# --- Matrix operations ---
W = np.random.randn(3, 5)  # weight matrix
x = np.random.randn(5, 1)  # input column vector
b = np.zeros((3, 1))       # bias vector

# Linear transformation: y = Wx + b
y = W @ x + b

print("W shape:", W.shape)
print("x shape:", x.shape)
print("y shape:", y.shape)
print("y =\n", y.round(3))
````

````python
# --- Broadcasting & reshaping ---
M = np.arange(12).reshape(3, 4)
row_vec = np.array([10, 20, 30, 40])

print("M:\n", M)
print("\nM + row_vec (broadcasting):\n", M + row_vec)
print("\nM flattened:", M.flatten())
````

````python
import matplotlib.pyplot as plt

# --- Quick plot to verify matplotlib ---
x = np.linspace(0, 2 * np.pi, 100)
plt.figure(figsize=(6, 3))
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)", linestyle="--")
plt.title("Quick Matplotlib Check")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()

print("If you see the plot above, matplotlib is working.")
````
