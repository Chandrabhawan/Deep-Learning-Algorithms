<summary><b>Solution 5: Optimizer Comparison</b></summary>

1. **SGD vs. SGD with Momentum:** Momentum accumulates a moving average of past gradients, which helps the optimizer move through narrow ravines and noisy gradients more smoothly. Without momentum, SGD can oscillate in directions with high curvature.

2. **Adagrad vs. Adam:** Adagrad accumulates the sum of all past squared gradients, so the effective learning rate monotonically decreases and can become vanishingly small. Adam fixes this by using exponentially decaying averages of both first and second moments, keeping the learning rate adaptive but bounded.

3. **When to choose SGD with momentum:** When you have compute budget for careful hyperparameter tuning and need the best generalization. SGD with momentum often generalizes better than Adam on image classification tasks and has been shown to find flatter minima.
</details>

<details>
<summary><b>Solution 7: Vanishing and Exploding Gradients</b></summary>

1. The sigmoid derivative is `sigmoid(z) * (1 - sigmoid(z))`, which has a maximum value of **0.25** (at z=0). When you multiply many values less than 1 through the chain rule across many layers, the gradient shrinks exponentially toward zero.

2. Three techniques to mitigate gradient problems:
   - **ReLU activation** (gradient is 1 for positive inputs)
   - **Residual/skip connections** (provide gradient highways)
   - **Proper weight initialization** (e.g., He or Xavier initialization)
   - (Also acceptable: gradient clipping, batch normalization, LSTM/GRU for RNNs)

3. **ReLU** largely solved the vanishing gradient problem because its derivative is 1 for all positive inputs, so gradients flow without shrinking through active neurons. Unlike sigmoid/tanh, it does not saturate for large positive values.
</details>

<summary><b>Solution 8: Debugging the Training Script</b></summary>

**Bug 1: Using sigmoid instead of ReLU for the hidden layer.**
Sigmoid causes vanishing gradients in deeper networks and saturates. For a hidden layer, ReLU is preferred. Fix: replace `sigmoid(z1)` with `np.maximum(0, z1)`.

**Bug 2: Missing softmax activation on the output layer.**
For multi-class classification, the output layer must produce a probability distribution over classes. The raw logits `z2` are returned directly. Fix: apply softmax -- `a2 = softmax(z2)`.

**Bonus issue:** MSE loss is not ideal for classification. Cross-entropy loss should be used with softmax outputs, as it provides stronger gradients when predictions are wrong and is theoretically grounded in maximum likelihood estimation.
</details>

<summary><b>Solution 9: 3-Layer Network Implementation</b></summary>

```python
import numpy as np

np.random.seed(0)
N = 150
X = np.random.randn(N, 4)
y_true = np.random.randint(0, 3, size=N)
Y = np.zeros((N, 3))
Y[np.arange(N), y_true] = 1

def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def cross_entropy_loss(y_pred, y_true_onehot):
    eps = 1e-12
    return -np.mean(np.sum(y_true_onehot * np.log(y_pred + eps), axis=1))

# He initialization
W1 = np.random.randn(4, 8) * np.sqrt(2.0 / 4)
b1 = np.zeros(8)
W2 = np.random.randn(8, 6) * np.sqrt(2.0 / 8)
b2 = np.zeros(6)
W3 = np.random.randn(6, 3) * np.sqrt(2.0 / 6)
b3 = np.zeros(3)

def forward(X):
    z1 = X @ W1 + b1
    a1 = relu(z1)
    z2 = a1 @ W2 + b2
    a2 = relu(z2)
    z3 = a2 @ W3 + b3
    a3 = softmax(z3)
    return {'z1': z1, 'a1': a1, 'z2': z2, 'a2': a2, 'z3': z3, 'a3': a3}

def backward(X, Y, cache):
    m = X.shape[0]
    a1, a2, a3 = cache['a1'], cache['a2'], cache['a3']
    z1, z2 = cache['z1'], cache['z2']
    
    dz3 = (a3 - Y) / m
    dW3 = a2.T @ dz3
    db3 = np.sum(dz3, axis=0)
    
    dz2 = (dz3 @ W3.T) * relu_derivative(z2)
    dW2 = a1.T @ dz2
    db2 = np.sum(dz2, axis=0)
    
    dz1 = (dz2 @ W2.T) * relu_derivative(z1)
    dW1 = X.T @ dz1
    db1 = np.sum(dz1, axis=0)
    
    return {'dW1': dW1, 'db1': db1, 'dW2': dW2, 'db2': db2, 'dW3': dW3, 'db3': db3}

lr = 0.01
for epoch in range(500):
    cache = forward(X)
    loss = cross_entropy_loss(cache['a3'], Y)
    grads = backward(X, Y, cache)
    
    W1 -= lr * grads['dW1']
    b1 -= lr * grads['db1']
    W2 -= lr * grads['dW2']
    b2 -= lr * grads['db2']
    W3 -= lr * grads['dW3']
    b3 -= lr * grads['db3']
    
    if epoch % 100 == 0:
        preds = np.argmax(cache['a3'], axis=1)
        acc = np.mean(preds == y_true)
        print(f"Epoch {epoch}: loss={loss:.4f}, accuracy={acc:.4f}")
```
