```python
x = torch.arange(24).reshape(2, 3, 4).float()

# 1. (2,3,4) -> (6,4) -- flatten first two dims
r1 = x.reshape(6, 4)           # or x.view(6, 4)
assert r1.shape == (6, 4)

# 2. (2,3,4) -> (2,12) -- flatten last two dims
r2 = x.reshape(2, 12)          # or x.view(2, -1)
assert r2.shape == (2, 12)

# 3. (2,3,4) -> (4,3,2) -- reverse dimensions
r3 = x.permute(2, 1, 0)        # NOT reshape! Order matters.
assert r3.shape == (4, 3, 2)

# 4. (2,3,4) -> (2,4,3) -- swap last two dims
r4 = x.permute(0, 2, 1)
assert r4.shape == (2, 4, 3)

# 5. (2,3,4) -> (2,3,2,2) -- split last dim
r5 = x.reshape(2, 3, 2, 2)     # or x.view(2, 3, 2, 2)
assert r5.shape == (2, 3, 2, 2)

print("All shapes correct!")
```

**Key insight:** `reshape`/`view` changes how data is grouped but preserves order. `permute` actually reorders the dimensions, which is needed for transpositions where element order changes.

```python
from torch.utils.data import TensorDataset, DataLoader

X = torch.randn(10, 4)
y = torch.randint(0, 3, (10,))

# 1. TensorDataset
dataset = TensorDataset(X, y)

# 2. DataLoader
loader = DataLoader(dataset, batch_size=3, shuffle=True)

# 3. Iterate
for i, (X_batch, y_batch) in enumerate(loader):
    print(f"Batch {i}: X={X_batch.shape}, y={y_batch.shape}")

# 4. Number of batches: ceil(10/3) = 4 batches
# Batches: 3, 3, 3, 1 (last batch has only 1 sample)
print(f"Number of batches: {len(loader)}")

# 5. drop_last=True
loader_drop = DataLoader(dataset, batch_size=3, shuffle=True, drop_last=True)
print(f"Batches with drop_last: {len(loader_drop)}")  # 3 batches
```

**Key insight:** With 10 samples and batch_size=3, the default DataLoader creates 4 batches (3+3+3+1). With `drop_last=True`, the incomplete final batch is dropped, leaving 3 batches of exactly 3 samples each.

<summary><b>Solution 7: Debugging -- Forgotten zero_grad</b></summary>

**What happens without `zero_grad()`:** PyTorch **accumulates** gradients by default. Each call to `loss.backward()` **adds** the new gradients to whatever is already in `.grad`. Without zeroing, gradients grow each iteration, causing erratic weight updates.

**Fixed loop:**

```python
torch.manual_seed(42)
model = nn.Linear(5, 1)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
X = torch.randn(20, 5)
y = torch.randn(20, 1)

for epoch in range(5):
    optimizer.zero_grad()  # FIXED: zero gradients before backward
    output = model(X)
    loss = loss_fn(output, y)
    loss.backward()
    optimizer.step()
    print(f"  Epoch {epoch+1}: loss = {loss.item():.4f}, "
          f"grad norm = {model.weight.grad.norm().item():.4f}")
```

You will see that the gradient norm is stable in the fixed version, while it grows without bound in the buggy version.
</details>

<summary><b>Solution 8: Complete Training Loop</b></summary>

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

torch.manual_seed(42)

# Data
X_train = torch.randn(200, 8)
y_train = (X_train[:, 0] + X_train[:, 1] > 0).float()
X_test = torch.randn(50, 8)
y_test = (X_test[:, 0] + X_test[:, 1] > 0).float()

# 1. Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 2. Model
model = nn.Sequential(
    nn.Linear(8, 16),
    nn.ReLU(),
    nn.Linear(16, 1),
    nn.Sigmoid(),
).to(device)

# 3. Loss and optimizer
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# 4. DataLoader
train_ds = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)

# 5. Training loop
for epoch in range(10):
    model.train()
    total_loss = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        output = model(X_batch).squeeze()
        loss = loss_fn(output, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}: avg loss = {total_loss/len(train_loader):.4f}")

# 6. Evaluation
model.eval()
with torch.no_grad():
    X_test_d = X_test.to(device)
    preds = (model(X_test_d).squeeze() > 0.5).float().cpu()
    accuracy = (preds == y_test).float().mean()
    print(f"Test accuracy: {accuracy:.4f}")
```
